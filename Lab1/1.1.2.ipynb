Libraries

```python
import numpy as np
import matplotlib.image as mpimg
from matplotlib import pyplot as plt
```

Read Image File

```python
image = mpimg.imread("../images/bird.jpg")
plt.imshow(image)
```

Transpose

```python
x_r = image[:,:,0]
y_r = x_r.transpose()
```

Show Results

```python
plt.figure(figsize=(12,8))

# แสดงผลแต่ละภาพ
plt.subplot(1, 4, 1)
plt.imshow(x_r, cmap='gray')
plt.title(f"X ({x_r.shape}, 3)")

plt.subplot(1, 4, 2)
plt.imshow(y_r, cmap='gray')
plt.title(f"Y (3, {y_r.shape})")

# ใช้ฟังก์ชัน moveaxis() เพื่อสลับแถวกับ channel สี เสมือนกับการทำ transpose
y_r_m = np.moveaxis(y_r, 0, -1)
plt.subplot(1, 4, 3)
plt.imshow(y_r_m, cmap='gray')
plt.title(f"Y (3, {y_r_m.shape})")

# ใช้ฟังก์ชัน reshape() เพื่อสลับมิติของ ความสูง กับ ความกว้าง
y_r_r = np.reshape(image, (image.shape[2], image.shape[0], image.shape[1]))
plt.subplot(1, 4, 4)
plt.imshow(y_r_r[0,:,:], cmap='gray')
plt.title(f"Y (3, {y_r_r.shape})")

plt.tight_layout()
plt.show()
```
