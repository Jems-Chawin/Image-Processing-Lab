Libraries

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

Read Image File

```python
image = cv2.imread("../images/bird.jpg")
image = image[:,:,[2,1,0]]
plt.imshow(image)
```

Quantization

```python
n = 4
q_level = 2**n

quantized_image = np.floor(image*(q_level/256))*(256/q_level)
quantized_image = quantized_image.astype(np.uint8)
```

Show

```python
plt.figure(figsize=(20,14))

plt.subplot(1,2,1)
plt.imshow(image, cmap="gray")
plt.title("Original Image")

plt.subplot(1,2,2)
plt.imshow(quantized_image, cmap="gray")
plt.title(f"Quatized Image (Bit Depth: {n})")

plt.tight_layout()
plt.show()
```
