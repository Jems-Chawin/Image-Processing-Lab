Libraries

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

Read Image File

```python
image = cv2.imread("../images/bird.jpg")
image = image[:,:,[2,1,0]]
plt.imshow(image)
```

```python
# Create a zeros array with the same dimensions as the original image
image_mask = np.zeros_like(image)

# Define the coordinates of the rectangle
x1, y1, x2, y2 = 200, 50, 400, 200

# Fill the rectangle region with white color
image_mask[y1:y2, x1:x2] = (255, 255, 255)
```

Show Results

```python
plt.figure(figsize=(20,14))

# Display the original image and the mask
plt.subplot(1,2,1)
plt.imshow(image_mask)

# Use Bitwise AND operation to mask only the object area from the original image
image_obj_area = cv2.bitwise_and(image, image_mask)

# Display the result (masked image)
plt.subplot(1,2,2)
plt.imshow(image_obj_area)

plt.tight_layout()
plt.show()
```
