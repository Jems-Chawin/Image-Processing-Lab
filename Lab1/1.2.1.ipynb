Libraries

```python
import cv2
import matplotlib.pyplot as plt
```

Read Image File

```python
img_bgr = cv2.imread("../images/bird.jpg")

img_rgb = cv2.cvtColor(img_bgr,cv2.COLOR_BGR2RGB)
img_hsv = cv2.cvtColor(img_rgb,cv2.COLOR_RGB2HSV)
img_hls = cv2.cvtColor(img_rgb,cv2.COLOR_RGB2HLS)
img_ycrcb = cv2.cvtColor(img_rgb,cv2.COLOR_RGB2YCrCb)

plt.figure(figsize=(20,14))

plt.subplot(1,5,1)
plt.imshow(img_bgr)
plt.title("BGR")

plt.subplot(1,5,2)
plt.imshow(img_rgb)
plt.title("RGB")

plt.subplot(1,5,3)
plt.imshow(img_hsv)
plt.title("HSV")

plt.subplot(1,5,4)
plt.imshow(img_hls)
plt.title("HLS")

plt.subplot(1,5,5)
plt.imshow(img_ycrcb)
plt.title("YCrCb")

plt.tight_layout()
plt.show()
```

Extract Each Channel of the Image Versions

```python
xr = img_rgb[:, :, 0]
xg = img_rgb[:, :, 1]
xb = img_rgb[:, :, 2]

yh = img_hsv[:, :, 0]
ys = img_hsv[:, :, 1]
yv = img_hsv[:, :, 2]

zh = img_hls[:, :, 0]
zl = img_hls[:, :, 1]
zs = img_hls[:, :, 2]

my = img_ycrcb[:, :, 0]
mcr = img_ycrcb[:, :, 1]
mcb = img_ycrcb[:, :, 2]
```

Show Results

```python
plt.figure(figsize=(14, 8))

plt.subplot(4, 4, 1)
plt.imshow(img_rgb)
plt.title("RGB")

plt.subplot(4, 4, 2)
plt.imshow(xr, cmap='gray')
plt.title("R")

plt.subplot(4, 4, 3)
plt.imshow(xg, cmap='gray')
plt.title("G")

plt.subplot(4, 4, 4)
plt.imshow(xb, cmap='gray')
plt.title("B")

plt.subplot(4, 4, 5)
plt.imshow(img_hsv)
plt.title("HSV")

plt.subplot(4, 4, 6)
plt.imshow(yh, cmap='gray')
plt.title("H")

plt.subplot(4, 4, 7)
plt.imshow(ys, cmap='gray')
plt.title("S")

plt.subplot(4, 4, 8)
plt.imshow(yv, cmap='gray')
plt.title("V")

plt.subplot(4,4,9)
plt.imshow(img_hls)
plt.title("HLS")

plt.subplot(4,4,10)
plt.imshow(zh, cmap='gray')
plt.title("H")

plt.subplot(4,4,11)
plt.imshow(zl, cmap='gray')
plt.title("L")

plt.subplot(4,4,12)
plt.imshow(zs, cmap='gray')
plt.title("S")

plt.subplot(4,4,13)
plt.imshow(img_ycrcb)
plt.title("YCrCb")

plt.subplot(4,4,14)
plt.imshow(my, cmap='gray')
plt.title("Y")

plt.subplot(4,4,15)
plt.imshow(mcr, cmap='gray')
plt.title("Cr")

plt.subplot(4,4,16)
plt.imshow(mcb, cmap='gray')
plt.title("Cb")

plt.tight_layout()
plt.show()
```
